```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

# PER
it = [1,2,3,4,5]
p = [0.88036809816,.919844861021,.931443638761,.93822131704,.932249322493]
r = [.887445887446,.880024737168,.873840445269,.854669140383,.850958565244]
f1 = [.883892824145,.899494310999,.901723037652,.894498381877,.88975105076]

sp = [.960208741031,.960208741031,.960208741031,.960208741031,.960208741031]
sr = [.924623115578,.924623115578,.924623115578,.924623115578,.924623115578]
sf = [.94208,.94208,.94208,.94208,.94208]

it2 = np.linspace(it[0], it[-1], 500)
p2 = interpolate.pchip_interpolate(it, p, it2)
r2 = interpolate.pchip_interpolate(it, r, it2)
f12 = interpolate.pchip_interpolate(it, f1, it2)

# plt.subplot(221)

plt.plot(it2, p2,"r",label='P')
plt.plot(it2, r2,"b",label="R")
plt.plot(it2, f12,"g",label="F1")
plt.plot(it, p, 'ro')
plt.plot(it, r, 'bs')
plt.plot(it, f1, 'g^')

plt.plot(it, sp, 'r--',label="SP")
plt.plot(it, sr, 'b--',label="SR")
plt.plot(it, sf, 'g--',label="SF1")


plt.legend(loc='upper left')
plt.title("PER on conll2003")

plt.xlim((1, 5))
plt.ylim((0.8, 1))
plt.xlabel('Iteration')
plt.ylabel('')


plt.show()
```
